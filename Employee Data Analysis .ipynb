**Handling missing values from data set;

```python
df.isnull().sum()
```

**Replace " missing value / null value " by empty string**

```python
df['Gender'].fillna('No Gender', inplace= True)
df
```

**After replacing the missing values of Gender column by empty string, we check again if there is any missing values or not**

```python
df.isnull().sum()
```

**Replacing all the missing values of Senior Management Column by calculating the MODE of that individual column**

```python
mode = df['Senior Management'].mode().values[0]

df['Senior Management'] = df['Senior Management'].replace(np.nan, mode)
```

```python
df
```

**So, after being replaced by MODE value from missing, we checked if there is anything as missing values or not**

```python
df.isnull().sum()
```

***Now we have to work on "First Name" and "Team" column.
As these are Arbitrary Data, so we cant fill it.
We have drop these these missing rows***

```python
df1 = df.dropna(axis=0, how='any')

print(df1.isnull().sum())
```

**We checked the whole data set after dropping all the missing / null values from the data set.     Before the data was " 1000,8 " and now data is " 899, 8"**

**Here we operated ENCODING OPERATION for the Gender value; Because categorical data can not be encoded untils its converting into NUMERICAL DATA**

**Visualizing data (Gender), as uni-variate analysis method, Method = HistPlot**

```python
sns.histplot(x='Salary', data=df) #binwidth=5000

plt.figure(figsize=(15, 4))
plt.show()
```

**Visualizing data (Gender), as bi-variate analysis method, Method = BoxPlot**

```python
sns.boxplot(x='Salary', y='Team', data=df)

plt.figure(figsize=(15, 4))
plt.show()
```

**Visualizing data (Gender AND Team), as bi-variate analysis method, Method = ScatterPlot**

```python
sns.scatterplot( x="Salary", y='Team', data=df, hue='Gender', size='Bonus %')

plt.legend(bbox_to_anchor=(1, 1), loc=2)
```

**For multivariate analysis, we can use pairplot()method of the seaborn module**

```python
sns.pairplot(df, hue='Gender', height=2)
```
